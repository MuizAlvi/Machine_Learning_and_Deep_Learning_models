# Data Augmentation

    Building new data based on modifications of existing data. This can help reduce overfitting.
    Our Modifications:
    
    • Flipping image (horizontally or vertically)
    • Rotating image
    • Zooming in or out of image
    • Cropping the image
    • Varying image codex

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc, ndimage
import keras
from keras import backend as k
from keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
# function taken from https://github.com/tensorflow/docs/blob/master/site/en/tutorials/images/classification.ipynb
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, 
    channel_shift_range=10., horizontal_flip=True)
```

```python
image_path = 'biddy.jpg'
```

```python
image = np.expand_dims(plt.imread(image_path),0)
plt.imshow(image[0])
```

```python
# Generate batches of augmented images
aug_iter = gen.flow(image)
```

```python
# Get 10 samples of augmented images
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
plotImages(aug_images)
```
